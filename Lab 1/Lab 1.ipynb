```python
def median(column_values):
    
    sorted_values = sorted(column_values) 
  
    n = len(sorted_values)
   
    if n % 2 == 1:
        median = sorted_values[n // 2]

    else:
        mid1 = sorted_values[n // 2 - 1]
        mid2 = sorted_values[n // 2]
        median = (mid1 + mid2) / 2
    
    return median
```

```python
def mode(df, attribute):
    # Drop NaN values
    values = df[attribute].dropna()
    
    # Dictionary to store the frequency of each value
    frequency_dict = {}
    
    # Iterate over the values and count frequencies
    for value in values:
        if value in frequency_dict:
            frequency_dict[value] += 1
        else:
            frequency_dict[value] = 1
    
    # Find the maximum frequency
    max_count = max(frequency_dict.values())
    
    # Collect all values that have the maximum frequency
    most_frequent_values = [value for value, count in frequency_dict.items() if count == max_count]
    
    return most_frequent_values
```

```python
def calculate_tendencies(df):
    
    #ep (ms)
    Ep_mode = mode(df, 'ep (ms)')

    #Acc_x    
    Acc_x_mean=  pd.to_numeric(df['Acc_x'], errors='coerce').mean()

    acc_x_values = pd.to_numeric(df['Acc_x'], errors='coerce').tolist() 
    Acc_x_median= median(acc_x_values)

    Acc_x_mode = mode(df, 'Acc_x')

    Acc_x_tendencies = [Acc_x_mean,Acc_x_median,Acc_x_mode]

    #Acc_y
    Acc_y_mode = mode(df, 'Acc_y')
    
    Acc_y_mean=  pd.to_numeric(df['Acc_y'], errors='coerce').mean()

    acc_y_values = pd.to_numeric(df['Acc_y'], errors='coerce').tolist() 
    Acc_y_median= median(acc_y_values)

    Acc_y_tendencies = [Acc_y_mean,Acc_y_median,Acc_y_mode]

    #Acc_z
    Acc_z_mode = mode(df, 'Acc_z')

    Acc_z_mean=  pd.to_numeric(df['Acc_z'], errors='coerce').mean()

    acc_z_values = pd.to_numeric(df['Acc_z'], errors='coerce').tolist() 
    Acc_z_median= median(acc_z_values)

    Acc_z_tendencies = [Acc_z_mean,Acc_z_median,Acc_z_mode]

    #Gyro_x
    Gyro_x_mean=  pd.to_numeric(df['Gyro_x'], errors='coerce').mean()

    gyro_x_values = pd.to_numeric(df['Gyro_x'], errors='coerce').tolist() 
    Gyro_x_median= median(gyro_x_values)

    Gyro_x_mode = mode(df, 'Gyro_x')

    Gyro_x_tendencies = [Gyro_x_mean,Gyro_x_median,Gyro_x_mode]

    #Gyro_y
    Gyro_y_mean=  pd.to_numeric(df['Gyro_y'], errors='coerce').mean()

    gyro_y_values = pd.to_numeric(df['Gyro_y'], errors='coerce').tolist() 
    Gyro_y_median= median(gyro_y_values)

    Gyro_y_mode = mode(df, 'Gyro_y')

    Gyro_y_tendencies = [Gyro_y_mean,Gyro_y_median,Gyro_y_mode]

    #Gyro_z
    Gyro_z_mean=  pd.to_numeric(df['Gyro_z'], errors='coerce').mean()

    gyro_z_values = pd.to_numeric(df['Gyro_z'], errors='coerce').tolist() 
    Gyro_z_median= median(gyro_z_values)

    Gyro_z_mode = mode(df, 'Gyro_z')

    Gyro_z_tendencies = [Gyro_z_mean,Gyro_z_median,Gyro_z_mode]
    
    #Label
    Label_mode = mode(df, 'Label')

    #Category
    Category_mode = mode(df, 'Category')

    #Set
    Set_mode = mode(df, 'Set')


    return Ep_mode, Acc_x_tendencies, Acc_y_tendencies, Acc_z_tendencies, Gyro_x_tendencies, Gyro_y_tendencies, Gyro_z_tendencies, Label_mode, Category_mode, Set_mode

   
```

```python
def calculate_quartile(df, attribute,mediane):
    
    # Drop NaN values
    values = pd.to_numeric(df[attribute], errors='coerce').dropna()

    sorted_values = sorted(values) 

    Q0 = sorted_values[0]
    
    #Q1 = sorted_values[len(sorted_values) // 4]

    Q1 = np.percentile(sorted_values, 25)

    Q2 = mediane

    #Q3 = sorted_values[ 3 * (len(sorted_values) // 4 )]

    Q3 = np.percentile(sorted_values, 75)

    Q4 = sorted_values[-1]

    return [Q0, Q1, Q2, Q3, Q4]
```

```python
def calculate_quartiles(df, medians):

    #Acc_x
    Acc_x_quartiles = calculate_quartile(df,"Acc_x",medians[0])

    #Acc_y
    Acc_y_quartiles = calculate_quartile(df,"Acc_y",medians[1])

    #Acc_z
    Acc_z_quartiles = calculate_quartile(df,"Acc_z",medians[2])

    #Gyro_x
    Gyro_x_quartiles = calculate_quartile(df,"Gyro_x",medians[3])

    #Gyro_y
    Gyro_y_quartiles = calculate_quartile(df,"Gyro_y",medians[4])

    #Gyro_z
    Gyro_z_quartiles = calculate_quartile(df,"Gyro_z",medians[5])

    return Acc_x_quartiles, Acc_y_quartiles, Acc_z_quartiles, Gyro_x_quartiles, Gyro_y_quartiles, Gyro_z_quartiles
```

```python
def display_missing_values(df, attribute):
    total_missing = df[attribute].isna().sum()
    percentage_missing = (total_missing / len(df)) * 100
    return total_missing, percentage_missing
```
