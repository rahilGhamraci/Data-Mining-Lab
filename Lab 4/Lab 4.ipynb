```python
def get_items_transactions(df):
    items = set()
    transactions = {}

    for _, row in df.iterrows():
        item = (row['Exercise'], row['Category'], row['Set'])
        items.add(item)

        if row['ID'] in transactions:
            transactions[row['ID']].add(item)
        else:
            transactions[row['ID']] = {item}

    return items, transactions
```

```python
from itertools import combinations 
def get_k_candidates(L_prev, k):
    """Generate candidate k-itemsets by joining L(k-1) with itself for k > 1, or return L_prev for k == 1."""
    if k == 1:
        return L_prev
    
    candidates = set() 

    for itemset1 in L_prev:
        for itemset2 in L_prev:
            if len(itemset1.union(itemset2)) == k and itemset1 != itemset2:   
      
                candidate = frozenset(itemset1.union(itemset2))
                candidates.add(candidate) 
    
    #purning 
    valid_candidates = set()
    for candidate in candidates:
        subsets_valid = True
        for subset in combinations(candidate, k-1):
            if frozenset(subset) not in L_prev:
                subsets_valid = False
                break
        if subsets_valid:
            valid_candidates.add(candidate)
    
    return list(valid_candidates)
```

```python
def calculate_support(df, C):
  
    supports = {}
    nb_transactions = len(df)


    for candidate in C:
        supports[frozenset(candidate)] = 0  
    
    for _, row in df.iterrows():
        items_set = row['Items']
        for candidate in C:
            if frozenset(candidate).issubset(items_set):
                supports[frozenset(candidate)] += 1
    

    for candidate in supports:
        supports[candidate] /= nb_transactions

    return supports
```

```python
supports = calculate_support(df2, candidates1)
for itemset, support in supports.items():
    print(f"Itemset: {set(itemset)}, Support: {support}")
```

```python
def generate_frequent_k_itemsets(df, L_prev, k, min_supp):
    """Generates the frequent k-itemsets from the given dataset and previous frequent itemsets (L_prev)."""
    frequent_itemsets = []  
    
    # generating candidates
    candidates = get_k_candidates(L_prev, k)

    # calculating supports
    supports = calculate_support(df, candidates)
  
    # deleting itemsets with supp < min_supp
    for itemset, support in supports.items():
        if support >= min_supp:
            frequent_itemsets.append(set(itemset)) 
    
    return frequent_itemsets
```

```python
def generate_association_rules(L):
    rules = []
    for itemset in L:

        itemset_list = list(itemset)
        for split_index in range(len(itemset_list)):  
            part1 = set(itemset_list[:split_index]) 
            part2 = set(itemset_list[split_index:]) 
            rule = (part1, part2)
            rules.append(rule)

    return rules
```

```python
rules = generate_association_rules(L5)
for rule in rules:
    print(f'{rule[0]} => {rule[1]}')
```

```python
def calculate_support2(df, itemsset):
  
    support = 0
    nb_transactions = len(df)
    
    for _, row in df.iterrows():
        items_set = row['Items']
        if itemsset.issubset(items_set):
            support += 1

    return support / nb_transactions
```

```python
def calculate_confidence(rule, df):
   
    if not rule[0]: 
        support_part2 = calculate_support2(df, rule[1])
        return support_part2
    
    support_part1 = calculate_support2(df, rule[0])
    support_part2 = calculate_support2(df, rule[0].union(rule[1]))
    
    return support_part2 / support_part1
```

```python
for rule in rules:
    confidence = calculate_confidence(rule,df2)
    print(f'confidence for {rule[0]} => {rule[1]}:\n{confidence}')
```
